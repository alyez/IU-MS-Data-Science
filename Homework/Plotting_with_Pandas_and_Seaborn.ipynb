```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
```

Seaborn comes with some datasets:

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('planets') #Load the planets data
```

```python
df.head()
```

### Why use seaborn?

Default matplotlib plot: not pretty

```python
plt.scatter(df.mass, df.distance)
```

Default seaborn plot: prettier

```python
sns.__version__
```

```python
sns.scatterplot(x = 'mass', y = 'distance', data = df)
```

### Some basic plots

`relplot` is the default option for visualizing the relationship between two numerical vatiables.

The default is scatter plot. We can also change to line plot:

```python
sns.relplot(x = 'mass', y = 'distance', data = df, kind="line")
```

As we've seen, Seaborn works with Pandas dataframes natually.

We can also use another column of df to specify the colors:

```python
sns.relplot(x = 'mass', y = 'distance', hue = "year", data = df)
```

Can we change the color palette?

```python
sns.relplot(x = 'mass', y = 'distance', hue = "year", data = df, palette = 'Paired')
```

List of available color palettes: https://seaborn.pydata.org/tutorial/aesthetics.html

Can we make the texts bigger?

```python
sns.set(font_scale=1.2)
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e"]
palette= sns.color_palette(flatui)
```

```python
sns.relplot(x = 'mass', y = 'distance', hue = "year", data = df, palette="viridis")
```

What if we want to use "year" as a categorical variable?

```python
df.year = df.year.astype("category") #Remember from last lab?
```

```python
sns.catplot(x="year", y="mass", data=df)
```

The x tick labels look no good...

**Here's one bad thing about seaborn:** It's often not straightforward to tweak the plot.
We have to go to the matplotlib object under the hood:

```python
p = sns.catplot(x="year", y="mass", data=df)
p.set_xticklabels(rotation=90)
```

We can also make it into a box plot.

```python
p = sns.boxplot(x="year", y="mass", data=df)
```

...but now the same trick for rotating the x tick labels doesn't work.

```python
p = sns.boxplot(x="year", y="mass", data=df)
p.set_xticklabels(rotation=90)
```

Instead:

```python
p = sns.boxplot(x="year", y="mass", data=df)
plt.xticks(rotation=90)
```

**Summary**: Seaborn is great for quickly making beautiful plots, but not if you want to have a lot of control.
