### More complicated operation
Add a column that says "fast car" if acceleration >= 12
and "slow car" if < 12

Use [pandas' apply](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.apply.html) to apply a function to each row

```python
def fast_slow(row):
    if row['Acceleration'] >= 12:
        return "fast car"
    else:
        return "slow car"
```

```python
df.apply(fast_slow, axis=1)
```

>axis : {0 or â€˜indexâ€™, 1 or â€˜columnsâ€™}, default 0<br>
Axis along which the function is applied:<br>
0 or â€˜indexâ€™: apply function to each column.<br>
1 or â€˜columnsâ€™: apply function to each row.<br>

```python
df['Fastness'] = df.apply(fast_slow, axis=1)
```

```python
df.head()
```

Simplied one-liner using Python's `lambda`

```python
df['Fastness'] = df.apply(lambda row:"fast car" if row['Acceleration'] >= 12 else "slow car", axis=1)
```
